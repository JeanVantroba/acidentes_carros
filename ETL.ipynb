Projeto para prática de ETL

```python
#importar biblioteca pandas
import pandas as pd
```

```python
#ler as planilhas de acidentes
df1 = pd.read_csv('acidentes2019_todas_causas_tipos.csv', encoding = "latin1", sep=';')
df2 = pd.read_csv('acidentes2020_todas_causas_tipos.csv', encoding = "latin1", sep=';')
df3 = pd.read_csv('acidentes2021_todas_causas_tipos.csv', encoding = "latin1", sep=';')
```

```python
#verifica as primeiras linhas de cada df
df1.head()
```

```python
df2.head()
```

```python
df3.head()
```

```python
#verifica as colunas para filtrar quais serão utilizadas no ETL

df1.columns
```

```python
#filtra as colunas que vamos usar para o processo ETL
df11 = df1[['id', 'data_inversa', 'uf','municipio','tipo_acidente', 'ano_fabricacao_veiculo']]
df22 = df2[['id', 'data_inversa', 'uf','municipio','tipo_acidente', 'ano_fabricacao_veiculo']]
df33 = df3[['id', 'data_inversa', 'uf','municipio','tipo_acidente', 'ano_fabricacao_veiculo']]
```

```python
#Verificar as colunas selecionadas
df11
```

```python
#junta os 3 dataframes em um só
df = pd.concat([df11, df22, df33])

df.head()
```

```python
df.columns
```

```python
#muda o nome das colunas
df = df.rename(columns={'data_inversa':'data'})
```

```python
df.columns
```

```python
#verifica a quantidade de linhas e colunas
df.shape
```

```python
#apresenta os estados com maior quantidade de acidentes
ranking_estado = df.groupby('uf')['id'].count().reset_index()
ranking_estado = ranking_estado.sort_values(by='id', ascending=False)

print(ranking_estado)
```

```python
#apresenta o ranking de acidentes por tipo de acidente
ranking_acidente = df.groupby('tipo_acidente')['id'].count().reset_index()
ranking_acidente = ranking_acidente.sort_values(by='id', ascending=False)

print(ranking_acidente)
```

```python
#apresenta o ranking de acidentes por ano de fabricação do carro
ranking_ano_veiculo = df.groupby('ano_fabricacao_veiculo')['id'].count().reset_index()
ranking_ano_veiculo = ranking_ano_veiculo.sort_values(by='id', ascending=False)

print(ranking_ano_veiculo)
```

```python
#salva os resultados em uma lista
nomeArquivo1 = 'ranking_acidentes1.xlsx'
nomeArquivo2 = 'ranking_acidentes2.xlsx'
nomeArquivo3 = 'ranking_acidentes3.xlsx'

ranking_acidente.to_excel(nomeArquivo1)
ranking_estado.to_excel(nomeArquivo2)
ranking_ano_veiculo.to_excel(nomeArquivo3)
```
